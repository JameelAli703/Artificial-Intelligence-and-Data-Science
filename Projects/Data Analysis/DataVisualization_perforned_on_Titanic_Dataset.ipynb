```python
# how to change the title of x and y axis in plot?
```

```python
# grid plot usinf facetgrid
g = sns.FacetGrid(kashti,  hue="survived",col='pclass', margin_titles=True,
              palette={1:"black", 0:"red"})
g = g.map(plt.scatter,"fare","age",edgecolor="w").add_legend()

```

```python
# grid plot usinf facetgrid
g = sns.FacetGrid(kashti,  hue="survived",col='who', margin_titles=True,
              palette={1:"seagreen", 0:"red"})
g = g.map(plt.scatter,"fare","age",edgecolor="w").add_legend()

```

```python
# grid plot usinf facetgrid
g = sns.FacetGrid(kashti,  hue="survived",col='sex', margin_titles=True,
              palette={1:"seagreen", 0:"red"})
g = g.map(plt.scatter,"fare","age",edgecolor="w").add_legend()
```

```python
#  heat map
# co relation heatmap of the dataset

corr = kashti.corr()
plt.figure(figsize=(10,10))
plt.title("Co relation Heat map of kashti dataset")
sns.heatmap(corr,annot=True,cmap="coolwarm")
```
